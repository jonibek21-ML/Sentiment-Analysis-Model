````python
data = [
    ("salom", "hello"),
    ("rahmat", "thank you"),
    ("men talabaman", "i am a student"),
    ("u kitob o'qiydi", "he reads a book"),
    ("biz ishlaymiz", "we work"),
    ("u aqldan ozgan", "he is stupid")
]
````

## 🔤 Tokenization
Tokenization — matnni model tushunadigan **raqamli indekslar ketma-ketligiga** aylantirish bosqichi.
Jarayon:
* matnni kichik harflarga o‘tkazish
* tokenlarga ajratish
* vocabulary yaratish
* token → index mapping
Misol:
```
hello world → [2, 5]
```
Maxsus tokenlar:
```
<start>  sentence start
<end>  sentence end
<padding>  padding
```
Bu bosqich Transformer pipeline’ning birinchi qadami hisoblanadi.

````python
def tokenize(data):
    return [(uz.split(), en.split()) for uz, en in data]
tokenized = tokenize(data)
def build_vocab(tokenized):
    uzb = {"<pad>": 0, "<start>": 1, "<end>": 2}
    eng = {"<pad>": 0, "<start>": 1, "<end>": 2}
    id_uz, id_en = 3, 3
    for uz, en in tokenized:
        for w in uz:
            if w not in uzb:
                uzb[w] = id_uz
                id_uz += 1
        for w in en:
            if w not in eng:
                eng[w] = id_en
                id_en += 1
    return uzb, eng

vocab_uz, vocab_en = build_vocab(tokenized)
````

````python
def index_pad(tokenized, vocab_uz, vocab_en):
    numeric = []
    for uz, en in tokenized:
        u_idx = [vocab_uz["<start>"]] + [vocab_uz[w] for w in uz] + [vocab_uz["<end>"]]
        e_idx = [vocab_en["<start>"]] + [vocab_en[w] for w in en] + [vocab_en["<end>"]]
        numeric.append((u_idx, e_idx))
    max_u = max(len(u) for u, _ in numeric)
    max_e = max(len(e) for _, e in numeric)
    padded = []
    for u, e in numeric:
        u_pad = u + [vocab_uz["<pad>"]] * (max_u - len(u))
        e_pad = e + [vocab_en["<pad>"]] * (max_e - len(e))
        padded.append((u_pad, e_pad))
    return padded

padded_data = index_pad(tokenized, vocab_uz, vocab_en)
src = torch.tensor([u for u, e in padded_data])
trg = torch.tensor([e for u, e in padded_data])
embed_size = 32
````

## 📍 Positional Encoding

Transformer tokenlar tartibini o‘zi bilmaydi.

Shuning uchun embeddinglarga positional encoding qo‘shiladi:

embedding + positional encoding

Bu modelga:

token joylashuvi
gap strukturasini

tushunishga yordam beradi.

Ushbu loyihada sinusoidal positional encoding ishlatilgan.

````python
embed_size = 32
def positional_encoding(seq_len, embed_dim):
    pe = torch.zeros(seq_len, embed_dim)
    for pos in range(seq_len):
        for i in range(embed_dim):
            if i % 2 == 0:
                pe[pos, i] = np.sin(pos / (10000 ** (i / embed_dim)))
            else:
                pe[pos, i] = np.cos(pos / (10000 ** ((i - 1) / embed_dim)))
    return pe
````

## 🏗 Encoder

Encoder kiruvchi gapni **contextual representation** ga o‘zgartiradi.
Encoder tarkibi:
```
Multi-Head Attention
Add & Norm
Feed Forward Network
Add & Norm
```
Natija:
```
source sentence → contextual embedding
```
Bu representation decoder uchun asosiy context vazifasini bajaradi.


````python
class EncoderBlock(nn.Module):
    def __init__(self, embed_size, heads):
        super().__init__()
        self.attn = nn.MultiheadAttention(embed_dim=embed_size, num_heads=heads, batch_first=True)
        self.ff = nn.Sequential(nn.Linear(embed_size, embed_size*4), nn.ReLU(), nn.Linear(embed_size*4, embed_size))
        self.norm1 = nn.LayerNorm(embed_size)
        self.norm2 = nn.LayerNorm(embed_size)

    def forward(self, x, src_key_padding_mask=None):
        attn_out, _ = self.attn(x, x, x, key_padding_mask=src_key_padding_mask)
        x = self.norm1(x + attn_out)
        x = self.norm2(x + self.ff(x))
        return x

class Encoder(nn.Module):
    def __init__(self, embed_size, heads, num_layers):
        super().__init__()
        self.layers = nn.ModuleList([EncoderBlock(embed_size, heads) for _ in range(num_layers)])

    def forward(self, x, src_key_padding_mask=None):
        for layer in self.layers:
            x = layer(x, src_key_padding_mask)
        return x
````

## 🏗 Decoder
Decoder tarjima gapini bosqichma-bosqich generatsiya qiladi.
Decoder tarkibi:
```
Masked Multi-Head Attention
Encoder-Decoder Attention
Feed Forward Network
```
Masked attention kelajak tokenlarni yashiradi,
encoder-decoder attention esa source context’dan foydalanadi.
Natijada model keyingi tokenni predict qiladi.


````python
class DecoderBlock(nn.Module):
    def __init__(self, embed_size, heads):
        super().__init__()
        self.self_attn = nn.MultiheadAttention(embed_dim=embed_size, num_heads=heads, batch_first=True)
        self.cross_attn = nn.MultiheadAttention(embed_dim=embed_size, num_heads=heads, batch_first=True)
        self.ff = nn.Sequential(nn.Linear(embed_size, embed_size*4), nn.ReLU(), nn.Linear(embed_size*4, embed_size))
        self.norm1 = nn.LayerNorm(embed_size)
        self.norm2 = nn.LayerNorm(embed_size)
        self.norm3 = nn.LayerNorm(embed_size)

    def forward(self, x, enc_out, tgt_mask=None, memory_key_padding_mask=None):
        self_attn, _ = self.self_attn(x, x, x, attn_mask=tgt_mask)
        x = self.norm1(x + self_attn)
        cross_attn, _ = self.cross_attn(x, enc_out, enc_out, key_padding_mask=memory_key_padding_mask)
        x = self.norm2(x + cross_attn)
        x = self.norm3(x + self.ff(x))
        return x

class Decoder(nn.Module):
    def __init__(self, embed_size, heads, num_layers):
        super().__init__()
        self.layers = nn.ModuleList([DecoderBlock(embed_size, heads) for _ in range(num_layers)])

    def forward(self, x, enc_out, tgt_mask=None, memory_key_padding_mask=None):
        for layer in self.layers:
            x = layer(x, enc_out, tgt_mask, memory_key_padding_mask)
        return x
````

## ⚙ Training
Training jarayonida model quyidagi pipeline orqali ishlaydi:
```
Tokenization → Embedding → Positional Encoding
→ Encoder → Decoder → Linear → Softmax
```
Loss function:
```
CrossEntropyLoss
```
Optimizer:
```
Adam
```
Training vaqtida **teacher forcing** usuli qo‘llanilgan.


````python
class Seq2Seq(nn.Module):
    def __init__(self, src_vocab_size, trg_vocab_size, embed_size, heads, num_layers):
        super().__init__()
        self.src_embed = nn.Embedding(src_vocab_size, embed_size, padding_idx=0)
        self.trg_embed = nn.Embedding(trg_vocab_size, embed_size, padding_idx=0)
        self.encoder = Encoder(embed_size, heads, num_layers)
        self.decoder = Decoder(embed_size, heads, num_layers)
        self.output_projection = nn.Linear(embed_size, trg_vocab_size)
        self.embed_size = embed_size

    def make_causal_mask(self, sz):
        """Upper-triangular mask to prevent decoder from seeing future tokens."""
        mask = torch.triu(torch.ones(sz, sz), diagonal=1).bool()
        return mask

    def forward(self, src, trg):
        src_pe = positional_encoding(src.shape[1], self.embed_size)
        trg_pe = positional_encoding(trg.shape[1], self.embed_size)

        src_emb = self.src_embed(src) + src_pe.unsqueeze(0)
        trg_emb = self.trg_embed(trg) + trg_pe.unsqueeze(0)

        src_pad_mask = (src == 0)   # True where padding
        tgt_causal_mask = self.make_causal_mask(trg.shape[1])

        enc_out = self.encoder(src_emb, src_key_padding_mask=src_pad_mask)
        dec_out = self.decoder(trg_emb, enc_out,
                               tgt_mask=tgt_causal_mask,
                               memory_key_padding_mask=src_pad_mask)

        logits = self.output_projection(dec_out)
        return logits
````

````python
model = Seq2Seq(
    src_vocab_size=len(vocab_uz),
    trg_vocab_size=len(vocab_en),
    embed_size=embed_size,
    heads=2,
    num_layers=2
)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss(ignore_index=0)

trg_input = trg[:, :-1]
trg_target = trg[:, 1:]

print("🚀 Training started...\n")
EPOCHS = 500
for epoch in range(1, EPOCHS + 1):
    model.train()
    optimizer.zero_grad()

    logits = model(src, trg_input)
    loss = criterion(logits.reshape(-1, len(vocab_en)), trg_target.reshape(-1))

    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
    optimizer.step()

    if epoch % 50 == 0:
        print(f"Epoch {epoch:4d} | Loss: {loss.item():.4f}")

print("\n✅ Training complete!\n")
````

## 🔍 Greedy Decoding
Inference vaqtida model tarjimani greedy decoding orqali generatsiya qiladi.
Jarayon:
```
<sos> → next token → next token → ... → <eos>
```
Har bosqichda:
```
argmax(probabilities)
```
tanlanadi.
Bu oddiy va tez ishlaydigan decoding algoritmi hisoblanadi.


````python
def greedy_decode(model, src_tensor, trg_vocab, max_len=10):
    model.eval()
    with torch.no_grad():
        src_pe = positional_encoding(src_tensor.shape[1], model.embed_size)
        src_emb = model.src_embed(src_tensor) + src_pe.unsqueeze(0)
        src_pad_mask = (src_tensor == 0)
        enc_out = model.encoder(src_emb, src_key_padding_mask=src_pad_mask)

        sos_id = trg_vocab["<start>"]
        eos_id = trg_vocab["<end>"]
        output_ids = [sos_id]

        for _ in range(max_len):
            dec_input = torch.tensor([output_ids], dtype=torch.long)
            trg_pe = positional_encoding(dec_input.shape[1], model.embed_size)
            dec_emb = model.trg_embed(dec_input) + trg_pe.unsqueeze(0)

            tgt_mask = model.make_causal_mask(dec_input.shape[1])
            dec_out = model.decoder(dec_emb, enc_out,
                                    tgt_mask=tgt_mask,
                                    memory_key_padding_mask=src_pad_mask)

            logits = model.output_projection(dec_out)
            next_id = logits.argmax(-1)[0, -1].item()
            output_ids.append(next_id)
            if next_id == eos_id:
                break

    inv_vocab = {v: k for k, v in trg_vocab.items()}
    tokens = [inv_vocab[i] for i in output_ids[1:] if inv_vocab[i] != "<end>"]
    return " ".join(tokens)
````
